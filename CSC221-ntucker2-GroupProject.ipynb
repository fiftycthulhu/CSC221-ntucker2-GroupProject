```python
import urllib.request
from bs4 import BeautifulSoup
import ssl
ssl._create_default_https_context = ssl._create_unverified_context
url = "https://en.wikipedia.org/wiki/World_War_II_casualties"
page = urllib.request.urlopen(url)
soup = BeautifulSoup(page, "lxml")
print(soup.prettify())
```

```python
# Look for all the <table...> tags on this page
all_tables = soup.find_all('table')
for table in all_tables:
    print("Table:" + str(table) + "/n")
```

```python
# Based on what was found in previousl search, narrow the search for 'wikitable sortable'
all_sortable_tables = soup.find_all('table', class_="wikitable sortable")

# Q: How many of these tables did we find?
print("NUMBER OF WIKITABLE SORTABLE TABLES FOUND:" + str(len(all_sortable_tables)))

for table in all_sortable_tables:
    print("Sortable Table:" + str(table) + "/n")
```

```python
# Found 3 'wikitable sortable' tables ... 
# We want the "Death toll of World War II & military wounded by country" table which is the first (index [0])
right_table = all_sortable_tables[0]
print("Right Table:" + str(right_table) + "/n")
```

```python
# Column save areas - LISTS
A = []  
B = []
C = []
D = []
E = []
F = []
G = []
H = []
I = []

# Get all the table rows
for row in right_table.find_all('tr'):
    # Get all the table data (cells) in each row
    cells = row.find_all('td')

    if len(cells) == 9:
        
        # COUNTRY
        # Country column has a flag & country link - Country name is link text (a.text)
        print("CELL[0] CONTENTS: " + str(cells[0]) + '\n')
        
        # NOTE: The text I want is either the link[0] or link[1]
        # When it is in link[1] then link[0] is always empty (''), so you need to test for ''
        country_link = cells[0].find_all('a')
        # REMOVE THIS LOOP - IT WAS ONLY NEEDED TO DIAGNOSE WHAT IS GOING ON
        if country_link[0].text == '':  
            country_name = country_link[1].text
        else:
            country_name = country_link[0].text
        print("COUNTRY NAME: " + country_name)
        print('\n')  
        A.append(country_name.strip())
        
        # TOTAL POPULATION
        if cells[1].find(text=True) == None :
            B.append("Blank")
        else:
            B.append(cells[1].find(text=True).strip())
            
        # MILITARY DEATHS
        if cells[2].find(text=True) == None :
            C.append("Blank")
        else:
            C.append(cells[2].find(text=True).strip())
            
        # CIVILIAN DEATHS - CAUSE MILITARY & CRIMES
        if cells[3].find(text=True) == None :
            D.append("Blank")
        else:
            D.append(cells[3].find(text=True).strip())
            
        # CIVILIAN DEATHS - CAUSE FAMINE & DISEASE            
        if cells[4].find(text=True) == None :
            E.append("Blank")
        else:
            E.append(cells[4].find(text=True).strip())
            
        # TOTAL DEATHS
        if cells[5].find(text=True) == None :
            F.append("Blank")
        else:
            F.append(cells[5].find(text=True).strip())
            
        # DEATHS AS % OF 1939 POPULATION
        if cells[6].find(text=True) == None :
            G.append("Blank")
        else:
            G.append(cells[6].find(text=True).strip())
            
        # AVG DEATHS AS % OF 1939 POPULATION
        if cells[7].find(text=True) == None :
            H.append("Blank")
        else:
            H.append(cells[7].find(text=True).strip())
            
        # MILITARY WOUNDED
        if cells[8].find(text=True) == None :
            I.append("Blank")
        else:
            I.append(cells[8].find(text=True).strip())  
```

```python
import pandas as pd

# Save our lists as columns in a Panda DataFrame
df = pd.DataFrame()
df['Country'] = A
df['Total Population 1/1/1939'] = B
df['Military deaths from all causes'] = C
df['Civilian deaths from military actions and crimes against humanity'] = D
df['Civilian deaths from war-related famine and disease'] = E
df['Total Deaths'] = F
df['Deaths as % of 1939 population'] = G
df['Average deaths as % of 1939 population'] = H
df['Military wounded'] = I
df
```
